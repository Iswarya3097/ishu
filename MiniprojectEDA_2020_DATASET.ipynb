# Categorical to categorical analysis

```python
df1=pd.crosstab(df['anaemia'],df['DEATH_EVENT'])
print(df1)
```

```python
observed_values=df1.values
print("observed values:\n",observed_values)
```

```python
val=stats.chi2_contingency(df1)
val
```

```python
expected_values=val[3]
print(expected_values)
```

```python
no_of_rows=len(df1.iloc[0:2,0])
no_of_cols=len(df1.iloc[0,0:2])
dof=(no_of_rows-1)*(no_of_cols-1)
print("Degree of freedom:",dof)
alpha=0.05
```

```python
from scipy.stats import chi2
chi_square=sum([(o-e)**2/e for o,e in zip(observed_values,expected_values)])
chi_square_statistic=chi_square[0]+chi_square[1]
p_value=1-chi2.cdf(x=chi_square_statistic,df=dof)
print("chi_square_statistic",chi_square_statistic)
print("p_value:",p_value)
print("significance level:",alpha)
print("Degree of freedom:",dof)
```

```python
if p_value<=alpha:
    print("Reject H0,There is relationship between two variables")
else:
    print("Retain H0, There is no relationship between two variables")
```

```python
#create independent and dependent features
columns=df.columns.tolist()
#filter the columns to remove the data we do not want
columns=[c for c in columns if c not in ['DEATH_EVENT']]
#store the target features
target='DEATH_EVENT'
#Define a random state
state=np.random.RandomState(42)
X=df[columns]
Y=df[target]
X_outliers=state.uniform(low=0,high=1,size=(X.shape[0],X.shape[1]))
#print the shape of x and y
print(X.shape)
print(Y.shape)

```

```python
#visualize the target variable
g=sns.countplot(df['DEATH_EVENT'])
g.set_xticklabels(['Alive','Dead'])
plt.show()
```

Here clearly see that there is a huge differences between the dataset.

```python
alive_count_0,dead_count_1=df['DEATH_EVENT'].value_counts()
alive_0=df[df['DEATH_EVENT']==0]
dead_1=df[df['DEATH_EVENT']==1]
print('alive_0:',alive_0.shape)
print('dead_1:',dead_1.shape)
```

```python
#Random under_sampling
alive_0_under=alive_0.sample(dead_count_1)
test_under=pd.concat([alive_0_under,dead_1],axis=0)
print("Total deathevent of 1 and 0:",test_under['DEATH_EVENT'].value_counts())
test_under['DEATH_EVENT'].value_counts().plot(kind='bar',title='count(target)')#plot the count after under_sampling
```

```python
#Random over_sampling
dead_1_over=dead_1.sample(alive_count_0,replace=True)
test_over=pd.concat([dead_1_over,alive_0],axis=0)
print("Total deathevent of 1 and 0:",test_over['DEATH_EVENT'].value_counts())
test_over['DEATH_EVENT'].value_counts().plot(kind='bar',title='count(target)')#plot the count after over_sampling
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
smote=SMOTE()
#fit predictor and target variable
x_smote,y_smote=smote.fit_resample(X,Y)
print('original dataset shape',Counter(Y))
print('Resample dataset shape',Counter(y_smote))
```

```python
from imblearn.under_sampling import NearMiss
from collections import Counter
nm=NearMiss()
#fit predictor and target variable
x_nm,y_nm=nm.fit_resample(X,Y)
print('original dataset shape',Counter(Y))

print('Resample dataset shape',Counter(y_nm))
```
